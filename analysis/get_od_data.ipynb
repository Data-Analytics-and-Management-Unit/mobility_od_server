```python
def format_coords(feature_list):
    res = ''
    
    for f in feature_list:
        coords = f['geometry']['coordinates']
        res += str(coords[1]) + ',' + str(coords[0]) + '|'

    return res[:-1]
```

```python
def split_list(l, chunk_size):
    for i in range(0, len(l), chunk_size):
        yield l[i: i + chunk_size]
```

```python
import copy
import requests
import json
from tqdm import tqdm

with open('data/bangalore_ward_centroid.geojson') as f:
    ward_centroid_points = json.load(f)

def get_od(origin, dest_list):
    params['origins'] = ''
    params['destinations'] = ''
    
    params['origins'] = format_coords([origin])
    params['destinations'] = format_coords(dest_list)

    res = requests.get(URL, params=params)

    return res
    
ward_p = ward_centroid_points['features']
od_matrix = []
for i in range(len(ward_p)):
    print(f'{i+1} of {len(ward_p)}')
    ward_c = copy.deepcopy(ward_p)
    origin = ward_c.pop(i)

    if origin['properties']['WARD_NO'] != '149':
        continue

    dest_list = list(split_list(ward_c, 25))

    dest_time_list = []
    dest_name_list = []
    for d in tqdm(dest_list):
        res = get_od(origin, d)
        
        if res.status_code == 200:
            json_res = res.json()
            dest_time_list += json_res['rows'][0]['elements']
            dest_name_list += json_res['destination_addresses']
        else:
            w_n = origin['properties']['WARD_NO']
            print(f'Issue with the response for {w_n}')

    od_matrix.append({
        'origin': origin,
        'dest': {
            'name_list': dest_name_list,
            'time_list': dest_time_list,
            'dest_list': ward_c
        }
    })

    with open('od_matrix_transit_149.json', 'w') as f:
        json.dump(od_matrix, f)
```
